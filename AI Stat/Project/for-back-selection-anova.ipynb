```python
import statsmodels.api as sm
import pandas as pd
from automation import frust_class, normalize_column
```

```python
data = pd.read_csv("HR_data.csv")
data["HighlyFrustrated"] = data["Frustrated"].apply(frust_class)
data = normalize_column(data, ["HR_Mean", "HR_Median", "HR_std", "HR_Min", "HR_Max", "HR_AUC"])
data
```

```python
formula = 'HighlyFrustrated ~ HR_Mean * HR_Median * HR_std * HR_Min * HR_Max * HR_AUC'
L = sm.OLS.from_formula(formula, data=data).fit()
anova = sm.stats.anova_lm(L, test='F', type=1).round(3)
pd.set_option('display.max_rows', None)
df = anova.sort_values("PR(>F)")
df
```
