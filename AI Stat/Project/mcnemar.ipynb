```python
import pandas as pd
from statsmodels.stats.contingency_tables import mcnemar 
```

```python
data = pd.read_csv("protected_group_predictions.csv")
data["Baseline"] = data["Baseline"] == data["True Values"]
data["Tree"] = data["Tree"] == data["True Values"]
data["KNN"] = data["KNN"] == data["True Values"]
data["LogReg"] = data["LogReg"] == data["True Values"]
data["ANN"] = data["ANN"] == data["True Values"]
data
```

```python
def mcnemar_test(col1, col2):
    print("For: ", col1, col2)
    col1_vals = data[col1].values
    col2_vals = data[col2].values
    contingency_table = pd.crosstab(col1_vals, col2_vals, margins=False)
    result = mcnemar(contingency_table, exact=False, correction=True)
    print("P-value:", result.pvalue, "\n")
```

```python
mcnemar_test("Baseline", "Tree")
mcnemar_test("Baseline", "KNN")
mcnemar_test("Baseline", "LogReg")
mcnemar_test("Baseline", "ANN")
```
